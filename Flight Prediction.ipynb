```python
train_data['Arrival_hour'] = pd.to_datetime(train_data['Arrival_Time']).dt.hour
train_data['Arrival_min'] = pd.to_datetime(train_data['Arrival_Time']).dt.minute
```

```python
train_data['Total_Stops'].value_counts()
```

```python
def Stops(x):
    if x == 'non-stop':
        return 0
    elif x == '1 stop':
        return 1
    elif x == '2 stops':
        return 2
    elif x == '3 stops':
        return 3
    elif x == '4 stops':
        return 4
```

```python
train_data['Total_Stops'] = train_data['Total_Stops'].apply(Stops)
```

```python
train_data.head()
```

```python
train_data['Airline'].value_counts()
```

```python
Airline = pd.get_dummies(train_data['Airline'], drop_first=True)
```

```python
Airline.head()
```

```python
Destination = pd.get_dummies(train_data['Destination'], drop_first=True)
```

```python
Destination.head()
```

```python
Source = pd.get_dummies(train_data['Source'], drop_first=True)
```

```python
Source.head()
```

```python
duration = list(train_data['Duration'])
```

```python
len(duration[2].split())
```

```python
for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + "0m"
        else:
            duration[i] = "0h" + duration[i]
```

```python
duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))
    
```

```python
min = int(duration[0].split("m")[0].split("h")[1])
```

```python
for i in range(len(duration)):
    min = int(duration[i].split("m")[0].split("h")[1])
    duration_mins.append(min)
```

```python
train_data['Duration_hours'] = duration_hours
train_data['Duration_minutes'] = duration_mins
```

```python
train_data.head()
```

```python
train_data.drop('Duration', axis = 1, inplace = True)
```

```python
data_train = pd.concat([train_data,Airline,Source,Destination], axis = 1)
```

```python
data_train.head()
```

```python
data_train.drop(['Airline', 'Source', 'Destination'], axis = 1, inplace = True)
```

```python
data_train.head()
```

```python
data_train.shape
```

```python
X = data_train.drop('Price', axis = 1)
```

```python
X.shape
```

```python
y  = data_train['Price']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
model = ExtraTreesRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_prediction = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_prediction)
```
